2: diz delle liste delle parole successive 

```python
file = open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r")
testo = file.read()
file.close()
testo = testo.replace("\n", " ")

specialChar = "!\"%()=?,.:;-"
#rimuovo i caratteri speciali

for char in specialChar:
    testo = testo.replace(char, " ")
testoDiviso = testo.split(" ")
```

```python
#rimuovere parole vuote e rendo case insensitive
testo1 = []
for parola in testoDiviso:
    if parola == "":
        continue
    testo1.append(parola.lower())
testoDiviso = testo1
testoDiviso
```

```python
frequenze = {}

for word in testoDiviso:    
    if word not in frequenze.keys():
        frequenze[word] = 1
    else:
        frequenze[word] += 1
frequenze
```

```python
#ricerca delle 10 parole più usate (COMPITO)
numParoleDaCercare = 10
max_freq = []

for i in range(numParoleDaCercare):
    max_freq.append((" ", 0))
    
#per tutte le parola di frequenze
for indice,parola in enumerate(frequenze):
    #questo è il valore della parola con frequenza relativa
    considerazione = (parola, frequenze[parola])
    
    #per tutte le parole salvate nella lista max frequenze
    for i in range(len(max_freq)):
        
        #se l'elemento ha la frequenza maggiore rispetto a quello della lista max_frequenze
        if considerazione[1] > max_freq[i][1]:
            
            #salvo il valore attuale della frequenza, ci metto il valore in consid e controllo per il valore prima scartato
            max_freq[i],considerazione = considerazione,max_freq[i]
            
print(max_freq)
```

```python
adiacenze = {}

for i in range(len(testoDiviso)-1):
    word = testoDiviso[i]
    succ = testoDiviso[i+1]
    
    #se la chiave non è presente nel dizionario
    if word not in adiacenze.keys():
        adiacenze[word]= []
        adiacenze[word].append(succ) 
    #controllo che la parola non sia stata gia inserita
    else:
        if succ not in adiacenze[word]:
            adiacenze[word].append(succ)
adiacenze
```
